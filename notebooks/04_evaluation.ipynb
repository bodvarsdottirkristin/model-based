# 04 – Evaluation

This notebook:
- Generates posterior predictive samples on the test set
- Computes predictive metrics (RMSE, MAE, R²)
- Plots predicted vs. actual values and residuals
- Saves figures to `results/figures/`

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.model import BayesianModel
from src.evaluation import evaluate_predictions
from src.visualization import plot_predictions, plot_residuals, save_figure
```

```python
# --- Demo: generate synthetic train/test data and fit the model ---
rng = np.random.default_rng(42)
n_train, n_test, p = 100, 30, 2
true_beta = np.array([2.0, -1.5])

X_train = rng.standard_normal((n_train, p))
y_train = X_train @ true_beta + rng.normal(0, 0.5, n_train)

X_test = rng.standard_normal((n_test, p))
y_test = X_test @ true_beta + rng.normal(0, 0.5, n_test)

model = BayesianModel(num_samples=1000, tune=500, chains=2, random_seed=42)
model.fit(X_train, y_train)
print('Model fitted.')
```

```python
# Generate predictions on the test set
y_pred = model.predict(X_test)
print('y_pred shape:', y_pred.shape)
```

```python
# Compute predictive metrics
metrics = evaluate_predictions(y_test, y_pred)
print('RMSE :', round(metrics['rmse'], 4))
print('MAE  :', round(metrics['mae'],  4))
print('R²   :', round(metrics['r2'],   4))
```

```python
# Predicted vs. actual
fig = plot_predictions(y_test, y_pred)
save_figure(fig, '../results/figures/predictions.png')
plt.show()
```

```python
# Residual plot
fig = plot_residuals(y_test, y_pred)
save_figure(fig, '../results/figures/residuals.png')
plt.show()
```
