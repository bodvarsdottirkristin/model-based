# 02 – Preprocessing

This notebook covers:
- Handling missing values
- Feature scaling
- Train / test split
- Saving processed data to `data/processed/`

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd

from src.data_loader import (
    load_csv,
    split_features_target,
    train_test_split_data,
    scale_features,
    handle_missing_values,
)
```

```python
# Load raw data – update path as needed
# df = load_csv('../data/raw/dataset.csv')

# --- Demo: synthetic data ---
rng = np.random.default_rng(42)
n = 200
df = pd.DataFrame({
    'x1': rng.standard_normal(n),
    'x2': rng.uniform(0, 10, n),
    'y':  rng.standard_normal(n) * 2 + 3,
})
# Introduce some missing values for demonstration
df.loc[rng.choice(n, 10, replace=False), 'x1'] = np.nan
print('Missing values before imputation:')
print(df.isnull().sum())
```

```python
# Handle missing values
df = handle_missing_values(df, strategy='mean')
print('Missing values after imputation:', df.isnull().sum().sum())
```

```python
# Split into features and target
X, y = split_features_target(df, target_col='y')
print('X shape:', X.shape, '  y shape:', y.shape)
```

```python
# Train / test split
X_train, X_test, y_train, y_test = train_test_split_data(X, y, test_size=0.2)
print('Train size:', X_train.shape[0], '  Test size:', X_test.shape[0])
```

```python
# Scale features
X_train_s, X_test_s, scaler = scale_features(X_train.values, X_test.values)
print('Train mean after scaling:', X_train_s.mean(axis=0).round(4))
print('Train std  after scaling:', X_train_s.std(axis=0).round(4))
```

```python
# Save processed splits (optional)
pd.DataFrame(X_train_s, columns=X.columns).assign(y=y_train.values).to_csv(
    '../data/processed/train.csv', index=False)
pd.DataFrame(X_test_s, columns=X.columns).assign(y=y_test.values).to_csv(
    '../data/processed/test.csv', index=False)
print('Processed data saved.')
```
