# 03 – Bayesian Model Definition and Training

This notebook:
- Loads the preprocessed training data
- Instantiates the `BayesianModel` (Bayesian linear regression via PyMC)
- Fits the model using NUTS MCMC
- Inspects posterior diagnostics (R-hat, ESS, trace plots)

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import arviz as az
import matplotlib.pyplot as plt

from src.model import BayesianModel
from src.inference import compute_diagnostics
from src.visualization import plot_trace, plot_posterior
```

```python
# Load preprocessed training data
# train = pd.read_csv('../data/processed/train.csv')
# X_train = train.drop(columns=['y']).values
# y_train = train['y'].values

# --- Demo: synthetic data ---
rng = np.random.default_rng(42)
n, p = 100, 2
true_beta = np.array([2.0, -1.5])
X_train = rng.standard_normal((n, p))
y_train = X_train @ true_beta + rng.normal(0, 0.5, n)
print('X_train shape:', X_train.shape)
```

```python
# Instantiate and fit the Bayesian model
model = BayesianModel(num_samples=1000, tune=500, chains=2, random_seed=42)
model.fit(X_train, y_train)
print('Model fitted successfully.')
```

```python
# Posterior summary
model.summary()
```

```python
# Convergence diagnostics
trace = model.sample_posterior()
diag = compute_diagnostics(trace)
print('Max R-hat    :', diag['rhat_max'])
print('Min ESS bulk :', diag['ess_bulk_min'])
print('Divergences  :', diag['divergences'])
```

```python
# Trace plots
fig = plot_trace(trace, var_names=['alpha', 'beta', 'sigma'])
plt.show()
```

```python
# Marginal posterior distributions
fig = plot_posterior(trace, var_names=['alpha', 'beta', 'sigma'])
plt.show()
```
