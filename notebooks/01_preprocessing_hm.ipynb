```python
import pandas as pd

df =  pd.read_csv(
            '/data/raw/pd_2018_present.csv', 
            dtype={
                18: "string",
                21: "string",
                26: "string",
            },parse_dates=["Incident Date"]
        )
```

```python
print(f"Before removal:")
print(df['Incident Date'].min(), df['Incident Date'].max())
print(df.shape)

df = df.drop(df[df['Incident Year'] == 2026].index)

print(f"\n")
print(f"After removal")
print(df['Incident Date'].min(), df['Incident Date'].max())
print(df.shape)
```

```python
df.head(10)
```

```python
df.columns
```

```python
df =    df[
            ['Row ID', 'Incident Number', 'Incident Datetime', 
             'Incident Date', 'Incident Time', 'Incident Year', 'Incident Code', 
             'Incident Category', 'Incident Subcategory',
             'Police District', 'Latitude', 'Longitude']
        ]
```

```python
print(" -------------------------- REPORT --------------------------")
print(f"Total number of incidents: ", df.shape[0])
print(f"Date range: ", df['Incident Date'].min(), "-", df['Incident Date'].max())
print(f"Number of incident categories: ", len(df['Incident Category'].unique()))
print(" -----------------------------------------------------------")
```

```python
df.to_csv("../data/processed/pd_2018_present.csv", index=False)
```

```python
# 3.2
# Upload the data
import matplotlib.pyplot as plt
# Total number of incidents per year
incidents_per_year = df['Incident Year'].value_counts().sort_index()

# Bar chart showing the incidents by year
incidents_per_year.plot(kind='bar', figsize=(12,6))
plt.ylabel('Number of Incidents')
plt.xlabel('Year')
plt.title('Total Number of Incidents per Year')

plt.tight_layout()
plt.show()
```

### Focus Crimes

Looking at all crimes together can obscure important patterns. Different types of crime may have very different trends.

Let's focus on a subset of 16 crimes — I've chosen these as examples of serious & common crimes:

```python
focus_crimes = [
  'Larceny Theft',
  'Assault',
  'Burglary',         
  'Motor Vehicle Theft',
  'Fraud',
  'Drug Offense',  
  'Missing Person',
  'Robbery',
  'Disorderly Conduct',
  'Weapons Offense',
  'Stolen Property',  
  'Arson',   
  'Traffic Collision',
  'Vandalism',
  'Sex Offense',
  'Prostitution',  
]
```

```python
df.columns
```

```python
df["Incident Datetime"] = pd.to_datetime(
    df["Incident Date"].astype(str) + " " + df["Incident Time"].astype(str),
    errors="coerce"
)
```

```python

df_plots = df[df['Incident Category'].isin(focus_crimes)].assign(Year=lambda x: x['Incident Datetime'].dt.year)
df_plots
```
